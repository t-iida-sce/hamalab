```python
def asc18acc_plots(df,save_name):
    c_dic = {"AE_Normal":"b", "VAE_Normal":"g", "CLS_CNNNet":"r"}
    m_dic = {"direct":"o", "addkd_affine":"^", "addkd_nonlin":"s"}
    plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%d'))

    fig, ax = plt.subplots(1,2,figsize=(8, 6))
    for i in range(len(df)):
        if df["vitlayer"][i]=='6':
            ax[0].scatter(x=int(df["guidedlayer"][i]),y=df["test_acc"][i],c=c_dic[df["t model"][i]],marker=m_dic[df["kd_layer"][i]], s=10)
    ax[0].plot([3,7],[0.863,0.863],linestyle = "dotted")
    ax[0].set_title('ViT of 6layers')
    ax[0].set_xlabel('guided_layer')
    ax[0].set_ylabel("Accuracy")
    ax[0].set_xticklabels(["2","3","4","5","6","7","8"])
    ax[0].set_ylim(0.78,0.88)
    ax[0].set_xlim(2,8)

    plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%d'))
    for i in range(len(df)):
        if df["vitlayer"][i]=='8':
            ax[1].scatter(x=int(df["guidedlayer"][i]),y=df["test_acc"][i],c=c_dic[df["t model"][i]],marker=m_dic[df["kd_layer"][i]], s=10)
    ax[1].plot([4.5,8.5],[0.863,0.863],linestyle = "dotted")
    ax[1].set_title('ViT of 8layers')
    ax[1].set_xlabel('guided_layer')
    ax[1].set_xticklabels(["4","5","6","7","8","9"])
    ax[1].set_ylim(0.78,0.88)
    ax[1].set_xlim(4,9)
    #legend_without_duplicate_labels(ax[1])
    plt.savefig(os.path.join(path,'PAT/{}_plots.jpg'.format(save_name)))
    plt.show()
def asc18f1_plots(df,save_name):
    c_dic = {"AE_Normal":"b", "VAE_Normal":"g", "CLS_CNNNet":"r"}
    m_dic = {"direct":"o", "addkd_affine":"^", "addkd_nonlin":"s"}
    fig, ax = plt.subplots(1,2,figsize=(8, 6))
    for i in range(len(df)):
        if df["vitlayer"][i]=='6':
            ax[0].scatter(x=int(df["guidedlayer"][i]),y=df["test_f1"][i],c=c_dic[df["t model"][i]],marker=m_dic[df["kd_layer"][i]], s=10)
    ax[0].plot([3,7],[0.622,0.622],linestyle = "dotted")
    ax[0].set_title('ViT of 6layers')
    ax[0].set_xticklabels(["2","3","4","5","6","7","8"])
    ax[0].set_xlabel('guided_layer')
    ax[0].set_ylabel("F1 score")
    ax[0].set_ylim(0.5,0.65)
    ax[0].set_xlim(2,8)

    for i in range(len(df)):
        if df["vitlayer"][i]=='8':
            ax[1].scatter(x=int(df["guidedlayer"][i]),y=df["test_f1"][i],c=c_dic[df["t model"][i]],marker=m_dic[df["kd_layer"][i]], s=10)
    ax[1].plot([4.5,8.5],[0.622,0.622],linestyle = "dotted")
    ax[1].set_title('ViT of 8layers')
    ax[1].set_xticklabels(["4","5","6","7","8","9"])
    ax[1].set_xlabel('guided_layer')
    ax[1].set_ylim(0.5,0.65)
    ax[1].set_xlim(4,9)
    plt.savefig(os.path.join(path,'PAT/{}_plots.jpg'.format(save_name)))
    plt.show()
```
