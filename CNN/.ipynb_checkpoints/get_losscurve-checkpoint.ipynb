```python
import os

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.metrics import confusion_matrix,f1_score,classification_report, accuracy_score,precision_score, recall_score
```

```python
def cnn_loss_dummy(dataset,model, layer,linear=0):

    data_name = ['lowfreq', 'shape','lowfreq2']
    model_list = ['CNN_flatten','CNNNET_flatten','MVCNN' ,'MVCNNNET',]
    linear_list = ['Linear','Nonlinear','pool']

    vis_path = './runs/{}/{}_CNN{}/each/{}'.format(data_name[dataset],model_list[model],layer,linear_list[linear])
    #vis_path = './runs_freq_{}/{}_ResNet/each/{}'.format(dataset,model_list[model],linear_list[linear])

    train_loss=np.array([])
    test_loss=np.array([])
    for p in range(5):
        path = os.path.join(vis_path,str(p),'classifier_train','his_loss.npy')
        train_loss = np.append(train_loss,np.load(path)[:,0])
        test_loss = np.append(test_loss,np.load(path)[:,1])

    train_loss = train_loss.reshape(5,300).T
    test_loss = test_loss.reshape(5,300).T

    print(train_loss.shape)
    print(test_loss.shape)

    train_loss_mean = np.mean(train_loss, axis=1)
    train_loss_std = np.std(train_loss, axis=1)
    test_loss_mean = np.mean(test_loss, axis=1)
    test_loss_std = np.std(test_loss, axis=1)

    plt.figure()
    plt.xlabel("Training examples")
    plt.ylabel("Loss")

    train_sizes=np.arange(300)

    # Traing score と Test score をプロット
    plt.plot(train_sizes, train_loss_mean, color="tab:blue", label="Train loss")
    plt.plot(train_sizes, test_loss_mean,color="tab:orange", label="Valid loss")

    # 標準偏差の範囲を色付け
    plt.fill_between(train_sizes, train_loss_mean - train_loss_std, train_loss_mean + train_loss_std, color="tab:blue", alpha=0.2)
    plt.fill_between(train_sizes, test_loss_mean - test_loss_std, test_loss_mean + test_loss_std, color="tab:orange", alpha=0.2)

    plt.legend(loc="best")
    plt.savefig(os.path.join(vis_path,'loss.png'))
    plt.show()

# model 0: CNN_flatten / 1: CNNNET_flatten / 2: MVCNN / 3: MVCNNNET
# layer 3 : CNN3(Convolution3) / 4 : CNN4(Convolution4) / 5 : CNN5(Convolution5) / 6 : CNN3(Convolution6)
for d in [0,1]:
    for m in [0,1,2,3]:
        for l in [4,5]:
            cnn_loss_dummy(dataset=d,model = m, layer = l)
```

```python
def cnn_loss_ecg(num_view,io,task,model,layer,cls_=0):

    input_order=['in_asc','in_rand']
    model_list = ['CNN_flatten','CNNNET_flatten','MVCNN' ,'MVCNNNET','MFB']
    cls_list = ['MLP','Linear','pool','poolmlp']

    vis_path = './runs/view{}_{}_{}/{}_CNN{}/each/{}'.format(num_view,task,input_order[io],model_list[model],layer,cls_list[cls_])
    #vis_path = './runs_freq_{}/{}_ResNet/each/{}'.format(dataset,model_list[model],linear_list[linear])

    train_loss=np.array([])
    test_loss=np.array([])
    for p in range(5):
        path = os.path.join(vis_path,str(p),'classifier_train','his_loss.npy')
        train_loss = np.append(train_loss,np.load(path)[:300,0])
        test_loss = np.append(test_loss,np.load(path)[:300,1])

    train_loss = train_loss.reshape(5,300).T
    test_loss = test_loss.reshape(5,300).T

    print(train_loss.shape)
    print(test_loss.shape)

    train_loss_mean = np.mean(train_loss, axis=1)
    train_loss_std = np.std(train_loss, axis=1)
    test_loss_mean = np.mean(test_loss, axis=1)
    test_loss_std = np.std(test_loss, axis=1)

    plt.figure()
    plt.xlabel("Training examples")
    plt.ylabel("Loss")

    train_sizes=np.arange(300)

    # Traing score と Test score をプロット
    plt.plot(train_sizes, train_loss_mean, color="tab:blue", label="Train loss")
    plt.plot(train_sizes, test_loss_mean,color="tab:orange", label="Valid loss")

    # 標準偏差の範囲を色付け
    plt.fill_between(train_sizes, train_loss_mean - train_loss_std, train_loss_mean + train_loss_std, color="tab:blue", alpha=0.2)
    plt.fill_between(train_sizes, test_loss_mean - test_loss_std, test_loss_mean + test_loss_std, color="tab:orange", alpha=0.2)

    plt.ylim(0,2)
    plt.legend(loc="best")
    plt.savefig(os.path.join(vis_path,'loss.png'))
    plt.show()

# model 0: CNN_flatten / 1: CNNNET_flatten / 2: MVCNN / 3: MVCNNNET
# layer 3 : CNN3(Convolution3) / 4 : CNN4(Convolution4) / 5 : CNN5(Convolution5) / 6 : CNN3(Convolution6)
for nv in [12,18]:
    for io in [0,1]:
        for task in [5]:
            for m in [4]:
                for l in [4]:
                    cnn_loss_ecg(num_view=nv,io=io,task=task,model = m, layer = l,cls_=1)
```

```python
def create_confusion_matrix(cm, labels,acc,f1,save_dir):
    sns.set()
    
    df = pd.DataFrame(cm)
    df.index = labels
    df.columns = labels

    if len(labels) > 4:
        char_size = 12
    else:
        char_size = 15

    fig, ax = plt.subplots(figsize=(8, 6))
    sns.set(font='IPAexGothic')
    sns.heatmap(df, annot=True,fmt='.3f',annot_kws={'size': char_size},linewidths=.5, ax=ax, cmap="YlGnBu",robust=True,vmin=0, vmax=1)
    plt.xlabel('Prediction')
    plt.ylabel('True')
    #plt.title('acc {:.3f}, f1 score {:.3f}'.format(acc,f1))
    plt.savefig('{}/cm.png'.format(save_dir))
    plt.show()
    plt.close()
```

```python
def cnn_acc(dataset,model, layer,linear=0):
    if dataset == 1:
        label_name = ['A','B','C','D']
    elif dataset == 0:
        label_name = ['A','B','C','D','E','F','G','H']
    elif dataset == 2:
        label_name = ['A','B','C','D','E','F']
        

    data_name = ['lowfreq', 'shape','lowfreq2']
    model_list = ['CNN_flatten','CNNNET_flatten','MVCNN' ,'MVCNNNET']
    linear_list = ['Linear','Nonlinear']

    vis_path = './runs/{}/{}_CNN{}/each/{}'.format(data_name[dataset],model_list[model],layer,linear_list[linear])
    #vis_path = './runs_freq_{}/{}_ResNet/each/{}'.format(dataset,model_list[model],linear_list[linear])

    test_acc=np.array([])
    for p in range(5):
        path = os.path.join(vis_path,str(p),'classifier_train','early_test_miss_patients.csv')
        df_acc_ = pd.read_csv(path,index_col=0)
        #test_acc = np.append(test_acc,)
        if p==0:
            df_acc = df_acc_
        else:
            df_acc = pd.concat([df_acc,df_acc_],axis=0)

    print(vis_path)
    targets = df_acc['target'].values
    predictions = df_acc['prediction'].values
    
    acc = accuracy_score(targets,predictions)
    f1 = f1_score(targets, predictions,average='macro')
    cm = confusion_matrix(targets, predictions,normalize='true')

    create_confusion_matrix(cm, label_name, acc, f1, vis_path)
    
# model 0: CNN_flatten / 1: CNNNET_flatten / 2: MVCNN / 3: MVCNNNET
# layer 3 : CNN3(Convolution3) / 4 : CNN4(Convolution4) / 5 : CNN5(Convolution5) / 6 : CNN3(Convolution6)

for d in [0,1]:
    for m in [0,1,2,3]:
        for l in [4,5]:
            cnn_acc(dataset=d,model = m, layer = l)
```
